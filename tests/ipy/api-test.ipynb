```python
import requests

def get_api_configurations(api_url, payload=None):
    try:
        
        # Enviar uma solicitação GET para a API
        response = requests.get(api_url, params=payload)

        # Exibir o código de status da resposta
        print(f"\n\nStatus Code: {response.status_code}")

        # Exibir os headers da API
        print("\nAPI Headers:")
        for header, value in response.headers.items():
            print(f"{header}: {value}")

        # Exibir o Content-Type suportado
        content_type = response.headers.get('Content-Type')
        print(f"\nContent-Type: {content_type}")

        # Exibir o Content-Length (tamanho da resposta)
        content_length = response.headers.get('Content-Length')
        if content_length:
            print(f"\nContent-Length: {content_length} bytes")
        else:
            print("\nContent-Length: Not provided")

        # Exibir Rate Limiting (se disponível)
        rate_limit = response.headers.get('X-RateLimit-Limit')
        rate_remaining = response.headers.get('X-RateLimit-Remaining')
        rate_reset = response.headers.get('X-RateLimit-Reset')

        print(f"\n\nX-RateLimit-Limit: {rate_limit}")
        print(f"X-RateLimit-Remaining: {rate_remaining}")
        print(f"X-RateLimit-Reset: {rate_reset}")

        # Exibir Retry-After (se disponível)
        retry_after = response.headers.get('Retry-After')
        print(f"\nRetry-After: {retry_after} seconds")

        # Exibir Last-Modified (se disponível)
        last_modified = response.headers.get('Last-Modified')
        print(f"\nLast-Modified: {last_modified}")

        # Exibir Content-Disposition (se disponível)
        content_disposition = response.headers.get('Content-Disposition')
        print(f"\nContent-Disposition: {content_disposition}")

        # Exibir o corpo da resposta (opcional)
        print("\nResponse Body:")
        print(response.json() if 'application/json' in content_type else response.text)

        # Exibir os headers da API
        print("\nAPI Headers:")
        for header, value in response.headers.items():
            print(f"{header}: {value}")


        print(f"\n\nFora do for:")
        # Exibir o Content-Type suportado
        content_type = response.headers.get('Content-Type')
        print(f"Content-Type: {content_type}")

        # Exibir o Content-Length (tamanho da resposta)
        content_length = response.headers.get('Content-Length')

        print(f"\nContent-Length: {content_length} bytes")


        # Exibir Rate Limiting (se disponível)
        rate_limit = response.headers.get('X-RateLimit-Limit')
        rate_remaining = response.headers.get('X-RateLimit-Remaining')
        rate_reset = response.headers.get('X-RateLimit-Reset')
        
        print(f"\n\nX-RateLimit-Limit: {rate_limit}")
        print(f"X-RateLimit-Remaining: {rate_remaining}")
        print(f"X-RateLimit-Reset: {rate_reset}")

        # Exibir Retry-After (se disponível)
        retry_after = response.headers.get('Retry-After')
        print(f"\nRetry-After: {retry_after} seconds")

        # Exibir Last-Modified (se disponível)
        last_modified = response.headers.get('Last-Modified')
        print(f"\nLast-Modified: {last_modified}")

        # Exibir Content-Disposition (se disponível)
        content_disposition = response.headers.get('Content-Disposition')
        print(f"\nContent-Disposition: {content_disposition}")

        # Exibir o corpo da resposta (opcional)
        print("\nResponse Body:")
        print(response.json() if 'application/json' in content_type else response.text)

    except requests.exceptions.RequestException as e:
        print(f"Erro ao conectar à API: {e}")

# URL da API e payload para teste
api_url = 'https://veiculos.fipe.org.br/api/veiculos/ConsultarTabelaDeReferencia'
# api_url = 'https://veiculos.fipe.org.br/api/veiculos/ConsultarMarcas'

payload = {
    "codigoTabelaReferencia": 312,
    "codigoTipoVeiculo": 1
}

# Chamar a função para obter as configurações
get_api_configurations(api_url, payload)
```
